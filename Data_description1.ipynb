```python
import pandas as pd

df = pd.read_csv("Player_Attributes_cleaned.csv", encoding="utf-8")
print(df.head())
```

```python
print("\nDeskriptive Statistik für overall_rating:")
print(df['overall_rating'].describe())

median_rating = df['overall_rating'].median()
print("\nMedian von overall_rating:", median_rating)
```

```python
# Histogramm zur Visualisierung der Verteilung von overall_rating
plt.figure(figsize=(8,6))
plt.hist(df['overall_rating'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Overall Rating')
plt.ylabel('Anzahl der Spieler')
plt.title('Histogramm: Verteilung der Gesamtbewertung')
# Zeichne eine rote gestrichelte Linie beim Median
plt.axvline(median_rating, color='red', linestyle='dashed', linewidth=2, label=f'Median: {median_rating}')
plt.legend()
plt.show()
```

```python
# Scatterplot, der den Zusammenhang zwischen "potential" und "overall_rating" zeigt
plt.figure(figsize=(8,6))
plt.scatter(df['potential'], df['overall_rating'], alpha=0.5)
plt.xlabel('Potential')
plt.ylabel('Overall Rating')
plt.title('Streudiagramm: Potential vs. Overall Rating')
plt.show()
```
